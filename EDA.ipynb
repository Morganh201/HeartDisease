```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, classification_report, mutual_info_score
from sklearn import metrics
```

```python
df = pd.read_csv("heart_train.csv")
df = df.drop('Unnamed: 0', axis=1)
df
```

### Univariate Analysis on Variables

```python
plt.title("Percentage of Heart Disease")
plt.pie([len(df) - df['HeartDisease'].sum(), df['HeartDisease'].sum()], labels = [0, 1], autopct='%1.1f%%')
print("Num of Heart Disease: ", df['HeartDisease'].sum())
print("Num of No Heart Disease: ", len(df) - df['HeartDisease'].sum())
plt.show()
```

#### Continuous Variables

```python
plt.figure(figsize=(12, 12))
cont_attributes = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
for i in range(len(cont_attributes)):
    plt.subplot(3, 2, i+1)
    plt.title(f"Distribution of {cont_attributes[i]}")
    plt.xlabel(cont_attributes[i])
    plt.ylabel("Num of ppl")
    plt.hist(df[cont_attributes[i]].loc[df['HeartDisease']==0], bins=40, alpha=0.5, label="No Heart Disease")
    plt.hist(df[cont_attributes[i]].loc[df['HeartDisease']==1], bins=40, alpha=0.5, label='Heart Disease')
    plt.legend()
plt.tight_layout()
plt.show()
```
