```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, classification_report, mutual_info_score
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('heart_train.csv')
df.head()
```

```python
# Target variable is Heart Disease, Binary Classification problem
df.info()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
# Change Sex to 1 or 0

df['Sex'] = df['Sex'].replace('M', 1)
df['Sex'] = df['Sex'].replace('F', 0)
df['ExerciseAngina'] = df['ExerciseAngina'].replace('Y', 1)
df['ExerciseAngina'] = df['ExerciseAngina'].replace('N', 0)
df
```

```python
X = df.drop(['HeartDisease', 'id'], axis=1)
y = df['HeartDisease']
```

```python
cat_variables = ['ChestPainType', 'RestingECG', 'ST_Slope']
def onehot(encoder, df, column_name):
    encoder_df = pd.DataFrame(encoder.fit_transform(df[[column_name]]).toarray())
    encoder_df = encoder_df.add_prefix(f'{column_name}_')
    df = df.join(encoder_df)
    df = df.drop(column_name, axis=1)
    return df

encoder = OneHotEncoder(handle_unknown='ignore')
le = LabelEncoder()
for attribute in cat_variables:
  X = onehot(encoder, X, attribute)
#X['ST_Slope'] = le.fit_transform(X['ST_Slope'])
X
```

```python
X_train, val_X, y_train, val_y = train_test_split(X, y, random_state=0)
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
predictions = lr.predict(val_X)
print(classification_report(predictions, val_y))

fpr, tpr, threshold = metrics.roc_curve(val_y, predictions)
auc = metrics.roc_auc_score(val_y, predictions)
plt.plot(fpr, tpr)
plt.title("ROC curve")
plt.ylabel("True Positive Rate")
plt.xlabel("False Positive Rate")
plt.show()

confusion_matrix = metrics.confusion_matrix(val_y, predictions)
display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[False, True])
display.plot()
plt.show()
```

```python
clf = MLPClassifier(max_iter=300).fit(X_train, y_train)
clf.fit(X_train, y_train)
predictions = clf.predict(val_X)
print(classification_report(predictions, val_y))

fpr, tpr, threshold = metrics.roc_curve(val_y, predictions)
auc = metrics.roc_auc_score(val_y, predictions)
plt.plot(fpr, tpr)
plt.title("ROC curve")
plt.ylabel("True Positive Rate")
plt.xlabel("False Positive Rate")
plt.show()

confusion_matrix = metrics.confusion_matrix(val_y, predictions)
display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[False, True])
display.plot()
plt.show()
```

```python
test_X = pd.read_csv('heart_test.csv')
id = test_X['id']
test_X = test_X.drop(['Unnamed: 0', 'id'], axis=1)

test_X['Sex'] = test_X['Sex'].replace('M', 1)
test_X['Sex'] = test_X['Sex'].replace('F', 0)
test_X['ExerciseAngina'] = test_X['ExerciseAngina'].replace('Y', 1)
test_X['ExerciseAngina'] = test_X['ExerciseAngina'].replace('N', 0)

for attribute in cat_variables:
  test_X = onehot(encoder, test_X, attribute)
#test_X['ST_Slope'] = le.fit_transform(test_X['ST_Slope'])

predictions = lr.predict(test_X)
predictions = pd.DataFrame({'id': id, 'output': predictions})
predictions = predictions.set_index('id')
submission = predictions.to_csv("submission.csv")
# index=False
```
