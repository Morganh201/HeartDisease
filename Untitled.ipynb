```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
import xgboost as xgb
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import RidgeClassifier
from sklearn.svm import SVC

from sklearn.metrics import mean_absolute_error, classification_report, mutual_info_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
df = pd.read_csv('heart_train.csv')
df = df.drop(['Unnamed: 0', 'id'], axis=1)
df.head()
```

```python
# Change Sex to 1 or 0

df['Sex'] = df['Sex'].replace({'M': 1, 'F': 0})
df['ChestPainType'] = df['ChestPainType'].replace({"TA": 1, "ATA": 2, "NAP": 3, "ASY": 4})
df['RestingECG'] = df['RestingECG'].replace({"Normal": 0, "ST": 1, "LVH": 2})
df['ExerciseAngina'] = df['ExerciseAngina'].replace({'N': 0, "Y": 1})
df['ST_Slope'] = df['ST_Slope'].replace({'Up': 1, "Flat": 2, "Down": 3})
df
```

```python
def outlier(attribute):
    # calculate IQR for column Height
    Q1 = df[attribute].quantile(0.25)
    Q3 = df[attribute].quantile(0.75)
    IQR = Q3 - Q1

    # identify outliers
    threshold = 1.5
    outliers = df[(df[attribute] < Q1 - threshold * IQR) | (df[attribute] > Q3 + threshold * IQR)]

    return outliers
```

```python
attributes = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
for attribute in attributes:
    outliers = outlier(attribute)
    df = df.drop(outliers.index, errors = 'ignore')
```

```python
X = df.drop(['HeartDisease'], axis=1)
y = df['HeartDisease']
```

```python
X.duplicated().sum()
```

```python
clf = LocalOutlierFactor(n_neighbors=10)
outlier_values = clf.fit_predict(X)
outlier_values.max()
```

```python
plt.bar([0, 1], y.value_counts())
plt.show()
```

```python
smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)
```

```python
scaler = StandardScaler()
X_res = scaler.fit_transform(X_res)
```

```python
X_train, val_X, y_train, val_y = train_test_split(X_res, y_res, test_size=0.3, random_state=2)
```

```python
def mlmodel(model, model_name, X_train, y_train, val_X, val_y):
    model.fit(X_train, y_train)
    predictions = model.predict(val_X)

    print(model_name)
    print(classification_report(predictions, val_y))
    fpr, tpr, threshold = metrics.roc_curve(val_y, predictions)
    auc = metrics.roc_auc_score(val_y, predictions)
    plt.plot(fpr, tpr)
    plt.title("ROC curve")
    plt.ylabel("True Positive Rate")
    plt.xlabel("False Positive Rate")
    plt.show()

    confusion_matrix = metrics.confusion_matrix(val_y, predictions)
    display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[False, True])
    display.plot()
    plt.show()
```

```python
rforest = RandomForestClassifier(random_state=2)
exforest = ExtraTreesClassifier(random_state=2)
reg = xgb.XGBClassifier(tree_method="hist", device="cuda", learning_rate=0.1, max_depth=7, subsample=0.7, random_state=42)
kneigh = KNeighborsClassifier(n_neighbors=10)
rneigh = RadiusNeighborsClassifier(radius=1)
gnb = GaussianNB()
lda = LinearDiscriminantAnalysis()
ridge = RidgeClassifier()
svm = SVC(random_state=0)
```

```python
models = [rforest, exforest, reg, kneigh, gnb, lda, ridge, svm]
for model in models:
    mlmodel(model, str(model), X_train, y_train, val_X, val_y)
```
